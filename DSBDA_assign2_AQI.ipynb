```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns


from sklearn.model_selection import train_test_split
from sklearn.preprocessing import PowerTransformer
from sklearn.linear_model import LinearRegression
from sklearn.tree import DecisionTreeRegressor
from sklearn.metrics import r2_score, mean_absolute_error, mean_squared_error
```

```python
data = pd.read_csv('AirQuality.csv', delimiter = ";")
```

```python
data
```

```python
#removing or dropping columns that are not required(Data Cleaning)
data.drop(columns=['Date', 'Time', 'Unnamed: 15', 'Unnamed: 16'], inplace= True)
```

```python
data
```

```python
#We now will replace commas with dots so the system recognizes them as float 
cols_to_fix = ['CO(GT)', 'C6H6(GT)', 'T', 'RH', 'AH']
for col in cols_to_fix:
    data[col] = data[col].str.replace(',' , '.')
    data[col] = data[col].astype(float)
```

```python
data
```

```python
#As we can see above now the data values have a period instead of the commas making the system able to read the values as float
```

```python
data = data.astype({
    'CO(GT)':'float64',
    'C6H6(GT)':'float64',
    'T':'float64',
    'RH':'float64',
    'AH':'float64'
})
```

```python
#Integration of data
data.dropna(inplace = True) #drops all the null values
data
```

```python
data.drop_duplicates(inplace = True, ignore_index = True)
```

```python
data
```

```python
# Data Transformation
def remove_outliers(data, column, lower =0.01, upper = 0.99):
    low = data[column].quantile(lower)
    high = data[column].quantile(upper) 
    return data[(data[column]>=low) & (data[column]<=high)]
```

```python
df = data.copy()

for col in df.columns:
    df = remove_outliers(df,col)
```

```python
df = df.reset_index(drop = True)
```

```python
pt = PowerTransformer(method= 'yeo-johnson')
df= pd.DataFrame(pt.fit_transform(df), columns = df.columns)
```

```python
#error corrections(Missing Values and Outliers)
print("Missing Values : \n", df.isnull().sum())
```

```python
print("Datatypes: \n", df.dtypes)
```

```python
X = df.drop ('AH', axis=1)
Y = df['AH']
X_train, X_test, y_train, y_test = train_test_split(
X, Y, train_size=0.8, random_state=6, shuffle=True)
```

```python
lr_model = LinearRegression ()
lr_model. fit(X_train, y_train)
lr_pred = lr_model.predict(X_test)
#LinearRegression model
```

```python
dt_model = DecisionTreeRegressor()
dt_model.fit(X_train, y_train)
dt_pred = dt_model.predict(X_test)
#DecisionTree Regressor
```

```python
def evaluate_model (name, y_true, y_pred):
    r2 = r2_score(y_true, y_pred)
    mae = mean_absolute_error(y_true, y_pred)
    mse = mean_squared_error(y_true, y_pred)


    print(f"\n{name} Model:")
    print (f"R2 Score: {r2:.4f}")
    print (f"MAE: {mae: .4f}")
    print(f"MSE: {mse:.4f}")
    print (f"Accuracy (%) : {r2*100:.2f}%")
```

```python
evaluate_model("Linear Regression", y_test, lr_pred)
evaluate_model("Decision Tree Regressor", y_test,dt_pred)
```
